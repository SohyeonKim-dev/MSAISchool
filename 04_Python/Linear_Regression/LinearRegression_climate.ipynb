# Scikit - Learn을 활용한 기후 변화 예측 모델

```python
# 1. numpy로 데이터 불러오기

import numpy as np

year, temp_diff = np.loadtxt("5-year-mean-1882-2014.csv", delimiter=',', usecols=(0,1), unpack = True)
```

```python
# 2. data 살펴보기

print(year[:9])
print(temp_diff[:9])
```

```python
# 3. data 시각화하기

import matplotlib.pyplot as plt

plt.scatter(year, temp_diff)
plt.title("scatter plot of temp diff vs year")
plt.xlabel("year")
plt.ylabel("temp deff")
plt.show()

# 매년 평균에 대한 온도차가 증가함을 확인할 수 있다. 
```

```python
# 4. Regression model 

from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(fit_intercept = True)

year = year.reshape(133,1)
linear_model.fit(year, temp_diff)
```

```python
# 5. model 예측

model_predict = linear_model.predict(year)
print(model_predict[:9])
```

```python
# 6. model 시각화

plt.scatter(year, temp_diff)
plt.plot(year, model_predict)
plt.title("real temp diff vs model predicted temp diff")
plt.show()
```

```python
# 7. model 확인

print('y={0}*x+{1}'.format(linear_model.coef_[0], linear_model.intercept_))
```
